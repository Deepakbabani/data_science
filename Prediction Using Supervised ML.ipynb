<h3>Prediction of Study Hours using Supervised ML <b>:</b></h3>

```
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
%matplotlib inline
url = "http://bit.ly/w-data" #(Dataset Link)
s_data = pd.read_csv(url)
print("Data imported successfully")
s_data.head(10)
```

```
# Plotting the distribution of Studied Hours 
s_data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours and Percentage')
plt.xlabel('Hours')  
plt.ylabel('Percentage')  
plt.show()
```

```
X = s_data.iloc[:, :-1].values  
y = s_data.iloc[:, 1].values  
print(X[0:10])
print(y[0:10])
```

```
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0) 
```

```
sns.barplot(x = 'Hours',y ='Scores',data = s_data, color='m')
```

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
line = regressor.coef_*X+regressor.intercept_
plt.scatter(X, y)
plt.title('Hours and Percentage')
plt.xlabel("Hours")
plt.ylabel("Percentage")
plt.plot(X, line)
plt.show()
```

```
# Predicting the scores
y_pred = regressor.predict(X_test)

# Comparing Actual vs Predicted data
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df
```

```
# pridicted score of a student if studied for 9.25 hours per day
hours = 9.25
own_pred = regressor.predict([[hours]])
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```
